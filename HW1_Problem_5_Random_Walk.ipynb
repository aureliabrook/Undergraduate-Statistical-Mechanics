```
######### Problem 5, part a

import random
import math
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('seaborn-whitegrid')

def steps(N, d = 2):
  numSteps = []
  for i in range(N):
    vec = []
    for j in range(d):
      vec.append(random.uniform(-0.5, 0.5))
    numSteps.append(vec)
  return numSteps
```

```
def XvsT(location):
  xVal = [location[0][0]]
  for i in range(1, len(location)):
    xVal.append(xVal[i-1]+location[i][0])
  
  plt.plot(list(range(len(location))), xVal)
  plt.title("Displacement over Time for a 10,000 Step Walk")
  plt.xlabel("Time")
  plt.ylabel("X Position")
  plt.show()
```

```
XvsT(steps(10000))
```

```
XvsT(steps(10000))
```

```
XvsT(steps(10000))
```

```
def XvsY(location):
  xVal = [location[0][0]]
  yVal = [location[0][1]]
  for i in range(1, len(location)):
    xVal.append(xVal[i-1]+location[i][0])
    yVal.append(yVal[i-1]+location[i][1])

  plt.plot(xVal, yVal)
  plt.title("Two Dimensional Random Walk of N Steps")
  plt.xlabel("X Position")
  plt.ylabel("Y Position")
  plt.show()
```

```
XvsY(steps(10))
```

```
XvsY(steps(1000))
```

```
XvsY(steps(100000))
```

The net distance increases by just under 10 for every jump in 100 steps, since $\sqrt{100} = 10$.

```
####### PART B

def endpoints(M, N):
  ends=[]
  for i in range(M):
    location = steps(N, 2)
    xVal = [location[0][0]]
    yVal = [location[0][1]]
    for i in range(1, len(location)):
      xVal.append(xVal[i-1]+location[i][0])
      yVal.append(yVal[i-1]+location[i][1])
    ends.append((xVal[-1], yVal[-1]))
  return ends
```

```
def distance(M, listN):
  for i in listN:
    ends = list(map(list, zip(*endpoints(M, i))))
    plt.plot(ends[0], ends[1], 'o')
  plt.legend(labels=(listN))
  plt.title("Endpoints of M Two-Dimensional Random Walks")
  plt.xlabel("X Position")
  plt.ylabel("Y Position")
  plt.show()
```

```
distance(10000,[10,1]) 
```

For the N = 1 case, the shape is a rectangle because it's bounded by ($\frac{-1}{2}$, $\frac{1}{2}$). The N = 10 case is noramlly distributed because of the central limit theorem. This makes it a circular shape of points that become more disperesed from each other as they increase away from 0. 

```
######### PART C

def displacement(M, listN):
  for n in listN:
    ends=[]
    for i in range(M):
      location = steps(n, 1)
      xVal = [location[0][0]]
      for j in range(1, len(location)):
        xVal.append(xVal[j-1]+location[j][0])
      ends.append(xVal[-1])
    plt.hist(ends, bins=50, histtype="bar")
    print("N= " + str(n)+" Mean: " + str(np.mean(ends)))
    print("N= " + str(n)+" Variance: " + str(np.var(ends)))
  plt.legend(labels=(listN))
  plt.title("Endpoints of M One-Dimensional Random Walks")
  plt.xlabel("Displacements")
  plt.ylabel("Counts")
  plt.show()
```

```
displacement(10000, [5, 3, 2, 1])
```

Going by percentages of the data within a section of the graph centered at zero, around 68% of the data exists within one standard devition (square root of the variance) of the mean for N = 5, which makes a Gaussian distribution is a very good approximation to the random walk. Therefore, a Gaussian becomes a good approximation very quickly.
